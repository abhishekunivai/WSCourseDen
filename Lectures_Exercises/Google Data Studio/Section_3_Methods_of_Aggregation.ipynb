### 3.2 Comparison Calculation

Now, in the above snippet, we can see that there are 3 more options after the different methods of aggregation. These are `Type`,`Comparison Calculation` and `Running Calculation`. 
* **Type :** This tells unit or data type of the data entered in the columns, like is it a number or a precentage or maybe a currency and so on.
* **Comparison Calculation :**  This option is used to display calculations while comparing it with other values. This is very useful in cases where we need to provide the changes in new data in comparison with the old data.

### 3.4 Drill Down

This is a very useful option to display  hierarchical data in a very compact manner. Let's say we have some related columns like region, country and city, then these can be shown in Drill Down mode.

The **`Drill Down`** option is available under the `Dimension` section in the `Data` Panel. 

image.png

After applying the `Drill Down` option, we see 2 arrows at the top right corner of the table. When you click on the downward pointing arrow, it goes a level deeper into the `Drill Down` i.e. in this case it will show Countries.



image.png

If we select a particular row and then click on the downward pointing arrow, then it will show the sub-categories of that particular column only.

**Example :** In this case if we select Europe and then click on the Drill Down arrow, then it will display only the European countries.
