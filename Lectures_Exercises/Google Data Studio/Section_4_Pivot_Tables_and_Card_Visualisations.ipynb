## **Section 4 : Pivot Tables and Card Visualisations**




### 4.1 Introduction to Pivot Tables and ScoreCards

Now that we have a decent idea of `Tables` in Google Data Studio in the previous section. In section, we'll work with **`Pivot Tables`**. Pivot Tables are used in situations which involve classification by rows as well as columns. So we'll dive into data details and have a look at key metrics and indicators for visualizations in reports.

### 4.2 Creating Pivot Tables

Let's talk about the different `Style` and `Data`options available with us in case of Pivot Tables other than what we generally have in normal Tables. Pivot Table enables us to organise data in multiple ways and visualize data better as it makes use of Rows, Columns and Metrics which open up a plethora of options for the user.

**Expand-Collapse :** This option is used to compactly display the dimensions of the pivot table and show a detailed view whenever necessary.

image.png

image.png

Enabling the `Expand-Collapse` option leads to addition of a "+" button on the top left Dimension. On clicking on this button, the Dimensions are expanded and all the Dimensions are visible.

image.png

A new option **`Highlight Color`**is available in the `Table Colors` section in styles.  

image.png

This option is basically used to provide a color to the column and row we want to focus on while presenting it to someone.

### 4.3 Comparison Calculations

Coming to comparison calculations, these are same as normal Tables. Here also we can aggregate our data by various criterias like `Sum`, `Average`,`Count`,`Min`,`Max`, etc.  

One thing to notice is that if we change the method of aggregation of our data, then the `Grand Total` also changes according to that particular method. 

image.png

### 4.4 Card Visualization and Interactive Filter

Scorecards are used to show a particular important performance indicator or metric. It highlights the particular indicator that is in focus and works as an eye-catcher.

image.png

You can simply create a Scorecard and then copy paste the same and change the `Method of Aggregation` to create scorecard displayed above.
Example: To create Average Sales, simply copy paste the Sales Scorecard and change its method of aggregation to Average instead of Sum.

Let's talk about **`Interactive Filter`**. This is a very useful option that is used to filter the data of other visualization tools on changing the filters of another visualization tool.



Example:
image.png

This is a table and some scorecards displaying various metrics from the table.

image.png

As soon as we click on computers, the scorecards also modify the data displayed and show the sales related only to the Product Category `Computer`.

To enable interactive filter, select the table and enable the **`Cross Filtering`** option available in the `Chart Interaction` section in `Data` panel.
