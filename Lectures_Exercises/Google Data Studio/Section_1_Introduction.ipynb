### 1.2 Google Data Studio v/s other softwares

**`Google Data Studio`**, **`Power BI`** and **`Tableau`** are the 3 most common Business Intelligence (BI) tools used for visulization of data to ease the understanding of numerics in business. These softwares have numerous powerful features like easy data import, data blending, filters, etc. Each of these software has it’s own advantages and disadvantages.

**Power BI**

Advantages:
1.	Allows numerous custom visualizations. 
2.	Easily blends with Microsoft Office and other platforms .

Cons:
1.	 Free and online versions have limited access to features.
2.	 Requires some pre-requisite skills like MS Excel for smooth working.


**Tableau**

Advantages:
1.	Provides suggestions of all the available visualizations for the given dimensions and values.
2.	 Provides Cloud-Assistance.
3.	 Easy to use and understand.

Cons:
1.	 Slightly slower in case of importing large datasets.
2.	 Slightly restricted of graphs and visualizations. 

**Google Data Studio**

Advantages:
1.	Free to use.
2.	Provides a huge basket of platforms to import data sources.
3.	Simple and Easy to Use

Cons:
1.	Based on what you’re attempting to make, there may be some knowledge trajectory.
2.	Links can sometimes be disconnected from third-party data sources, which ends in data being missed.
