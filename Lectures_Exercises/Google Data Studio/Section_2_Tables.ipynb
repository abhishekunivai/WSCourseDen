### 2.1 Introduction to Tables

Tables are a great tool to summarize our data. It allows us use different methods to aggregate the data, like Sum, Min, Max etc. 
We'll be formatting and styling different tables and also learn about **`conditional formatting` , `cross filtering`** etc. We'll also have a look at different calculations like **`percentage calculations`**, running **`total calculations`**. So, Let's begin the show! 


### 2.2 Creating a Table

To create a Table, click on the **`Add a chart`** button present in the centre of the ribbon below the Menu bar and select **`Table`**. There are other options available as well which we'll discuss later. 
Left click on the location where you want to insert a table. 

side_panel.png

side_panel2.png


After creating the table, a new tab pops-up on the right side of the Report. It consists of these fields:
* **Data Source :** This field contains the source from which the data in the table is displayed. You can add more data into the table by clicking on **`Blend Data `** and selecting the desired data source.
* **Data Range Dimension :** This is used when you want to Filter your data and you want to apply the Filter to different data ranges. We'll talk more on this in the coming sections.
* **Dimension :** This field basically contains the text columns of your Table, or you may say as the x-axis if the table is plotted. We can change the default dimension by simply dragging the desired column from the list of available columns in the right panel.
* **Metric :** This field contains all the columns that you want to display in your table alongwith the dimension.

* **Sorting :** This field is used to sort the desired columns based on our preferences.

We can also add a `Summary Row` at the bottom which will display aggregate data of respective column.


### 2.3 Formatting and Sorting Tables

Now, we'll look at formatting options availble for tables. Click on the **`Style`** bar available on the top of the right ribbon alongside the Data button. The panel appears as follows:

Style_Bar.png

We'll talk about conditional formatting later in this section. Following that are the options of `Table Header` and `Wrap Text`. Wrap Text adjusts the text of a column such that the whole text is visible even if the available width of the column is less. 

**Table Header :** One can change the header font color, header font size and header font family as well.

**Table Colors :** You can change Header Background, Cell Border, Odd Row and Even row colors in this section.

**Table Footer :** Here, we can show/hide/compact pagination. Pagination is the bottom-most line of a table that displays some details about the table like how many rows are currently displayed and many more.


There are numerous other sections available as well like `Table Body`, `Table Labels`, `Missing Data`, `Dimension Allignment`, `Background and Border` etc. Try working with these sections and see how it changes the formatting of the table. 





**Sorting :** There is an arrow along-with the column name in the table, you can use that arrow to sort the data of the table according to that column, be it in Ascending or Descending order.

### 2.4 Multiple Measures

Let's talk about the concept of **`Optional Metrics`**. By turning on optional metrics, we can hide the unwanted columns for an overview, and show them when going for a detailed review.

This option is available under the `Metrics` section in the Data Panel. After turning on the toggle bar for optional metrics, a new icon appears on the top right corner of the table. We can select the desired columns from this icon.

### 2.5 Conditional Formatting

Other than numbers, we can also show the numerical values of a table in the form of **`Heatmaps`** or **`Bars`**. The color of the heatmap can be chosen as one wants and in case of bar, we can also setup a target line along with an axis which tells us about which particular rows have achieved the respective target or not.

image.png

Coming to **`Conditional Formatting`**, it means that we need to apply formatting only to those rows/cells/columns that fulfil some applied condition.

Let's try to mark all the cells in Col4 as Green if they are more than 20.

In the Style panel, click on the **`Add`** button under the `Conditional Formatting` section. Now, select Color Type and the column to which we want to apply the color. Select the condition and respective color and style and click on `Save`.
