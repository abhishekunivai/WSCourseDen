###### Select

Command : SELECT field_name1, field_name 2,... field_nameN   
FROM table_name1, table_name2...  
[WHERE condition]  
[GROUP BY field_name(s)]  
[HAVING condition]   
[ORDER BY field_name(s)]  
[OFFSET M ][LIMIT N];

Use * for all columns/fields.
