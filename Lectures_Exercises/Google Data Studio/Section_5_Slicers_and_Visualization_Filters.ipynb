## **Section 5 : Slicers and Visualization Filters**

### 5.1 Introduction to Slicers and Visualization Filters

Till now we've seen that we can filter data that is provided to us with the help of visualization tools like `Tables`, `Pivot Tables` etc. In this section, we'll look at using `Text`, `Date`, `Numbers` as filters. These options are very useful in cases where we want to see the data of past 10 days or if we want to chain filters. Another advantage of these filters is that we can apply a visual filter and then further add-on another filter on it as well. This makes it a very powerful tool.

### 5.3 Advanced Filter

Go to `Add a Control` and select **`Advanced Filter`**. This control has a variety of filters to classify the data like `Equals`,`Contains`, `Starts with`,`Regex` and `In`.  

image.png

**Equals (default) :** The value exactly matches the search term.

**Contains :** The value contains the search term.

**Starts With :** The value begins with the search term.

**Regex :** The value matches the regular expression

**In :** The value matches one or more of the search terms (enter search terms separated by commas).


In the example below, we have selected `Starts with` 'i' in the advanced filter and it's visible in the `Fixed-Sized List` that only countries starting with i are visible.

### 5.4 Numeric Slicers

In this section, we'll talk about **`Slider`** control. Here's a look of the slider control.

image.png

image.png

This is how the `Style` panel looks for the Slider control.
The `Slider mode` offers the following options:

**Range :** Makes the minimum and maximum values adjustable.

**Single value :** Lets you use the slider to select a single value.

**Locked min :** Makes only the maximum value in the range adjustable.

**Locked max :** Makes only the minimum value in the range adjustable.

**`Snapping`**

**Step size :** Sets the size of each movement of the slider thumb. For example, to have each move of the slider thumb increase or decrease by 10, set Step size to 10.

**Show tick marks :** Shows marks on the slider bar at the specified interval.
